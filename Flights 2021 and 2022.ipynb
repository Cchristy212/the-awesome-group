```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Create paths to csv files
flights2021_path = 'Combined_Flights_2021.csv'
flights2022_path = 'Combined_Flights_2022.csv'
```

```python
# Read the csv files = pd.read_csv(flights 2021 and 2022)
flights2021_data = pd.read_csv(flights2021_path)
flights2022_data = pd.read_csv(flights2022_path)
```

```python
flights2021_data.columns
```

```python
flights2021_data.Airline.value_counts()
```

```python
#Filtering the data to just two airlines Hawaiian Airlines Inc and Horizon Air

key1 = flights2021_data.Airline == "Hawaiian Airlines Inc."
key2 = flights2021_data.Airline == "Horizon Air"

filter_airline21 = flights2021_data[key1 | key2]

#filter_dest_21 = flights2021_data[flights2021_data.Dest == ["MIA" , "PHX"]]

filter_airline21
```

```python
#Filtering the data to just two airlines Hawaiian Airlines Inc and Horizon Air

key1 = flights2022_data.Airline == "Hawaiian Airlines Inc."
key2 = flights2022_data.Airline == "Horizon Air"

filter_airline22 = flights2022_data[key1 | key2]

#filter_dest_22 = flights2022_data[flights2021_data.Dest == ["MIA" , "PHX"]]

filter_airline22
```

```python
df = pd.merge(filter_airline21, filter_airline22, how="left", on= "Airline")
```

```python
# Combine the data into a single dataset
df = pd.merge(flights2021_data, flights2022_data, how="left", on="FlightDate")

# Display the data table for preview
df
```

```python

#check if there are any duplicates and drop them 

# making a bool series
flights_dup = df.duplicated(keep = False)
 
# bool series
flights_dup
 
# passing NOT of bool series to see unique values only
raw_data_df1 = df[~flights_dup]
 
# displaying data
raw_data_df1.info()
raw_data_df1
```
